```python
df['log_target'] = np.log1p(df['total_recovery_cost'])

sns.histplot(df['log_target'])
```

```python
low_cost = df[df['log_target'] < 10][['disasterNumber', 'incident_type', 'states_affected', 'total_recovery_cost']]
print(low_cost)
```

```python
plt.figure(figsize=(8, 8)) 
fig, axes = plt.subplots(1, 2, figsize=(18, 8))

# 1. Incident type counts
sns.countplot(data=df, y='incident_type', order=df['incident_type'].value_counts().index, ax=axes[0])

# 3. Incident type vs log_target
sns.boxplot(data=df, x='log_target', y='incident_type', ax=axes[1])

plt.tight_layout()
plt.show()
```
